```python
from bs4 import BeautifulSoup
import webscrap.io
import webscrap.html
```

앨런 튜링 위키백과 스크랩

https://ko.wikipedia.org/wiki/%EC%95%A8%EB%9F%B0_%ED%8A%9C%EB%A7%81

스냅샷 파일 생성

```python
url = 'https://ko.wikipedia.org/wiki/%EC%95%A8%EB%9F%B0_%ED%8A%9C%EB%A7%81'
웹문서스냅샷파일 = webscrap.io.snapshot_webpage(url, '앨런튜링_위키.html')
```

BeautifulSoup 생성

```python
html = open(웹문서스냅샷파일)
soup = BeautifulSoup(html, 'html.parser')
```

제목 추출

```python
soup.title
```

본문 추출

```python
본문목록 = soup.select('p')

for 본문 in 본문목록:
    print(본문.text)
```

표 추출

```python
표목록 = soup.select('table.wikitable')
webscrap.html.표추출(표목록)[0]
```
